# Probability Density

Some dependencies

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

## Definition

The probability density $p)x)$ is a function that describes the probability of a continous variable $x$ in a specfic value.

However, we the probability of a continuos variable take a exactly value is 0, but why? this is because $p(x)$ describe a "density". This density its not a probability itself, but is a measure of how likely is that the variable is near that point.

For continous variables the probability of the variable take a exact values is 0 because there are infinite possible values that the variable cante take. In a continous space, assign a non-zero probability to an exact value would imply ignoring the infinite number of other possible values.

We define the probability density $p(x)$ over a continuous variable of $x$ falling in the interval($x$,$x$+ $\delta x)$ is given by $p(x)$ $\delta x$ for $\delta x \rightarrow 0$

Interval ($x$,$x$+ $\delta x)$: This is a very small interval around $x$

The probability of the variable $x$ falls in the interval ($x$,$x$+ $\delta x) can be approximated as the product of $p(x) \delta x$:

$P(x≤X≤x+δx)≈p(x)⋅δx$

As $\delta x$ tends to 0, $\delta x \rightarrow 0$ this approximation becomes more accurate. So far, we can understand that this is approximation is in order to get an a approximation of get accurate result of $x$ in $p(x)$.

The probability that $x$ will lie in an interval $(a,b)$ is then given by:

$p(x \in (a, b)) = \int_{a}^{b} p(x) \, dx$

Also because probabilities are none negative, and because the value of x must lite some-where on the real axis, p(x) must satisfy two conditions:

$p(x) \geq 0$

$\int_{-\infty}^{\infty} p(x) \, dx = 1$

The probability that x lies in the interval $(-\infty, z)$ is given by the cumulative distribution function defined by:

$P(z) = \int_{-\infty}^{z} p(x) \, dx$

which satisfies $P'(x)=p(x)$

### Code example of the definition

```python
# definition of p(x) as a gaussian normal distribution
def p(x):
    return norm.pdf(x, loc=0, scale=1)

# x range
x = np.linspace(-3, 3, 1000)
px = p(x)

#  acumulated probability
Px = norm.cdf(x, loc=0, scale=1)

# define interval of delta x
delta_x = 0.2
x_interval_start = 0.5
x_interval_end = x_interval_start + delta_x

# plots
fig, ax1= plt.subplots(figsize=(10, 6))

# p(x)
ax1.plot(x, px, label='$p(x)$', color='red')
ax1.fill_between(x, px, where=(x >= x_interval_start) & (x <= x_interval_end), color='green', alpha=0.5)
ax1.set_xlabel('$x$')
ax1.set_ylabel('$p(x)$', color='red')
ax1.tick_params(axis='y', labelcolor='red')
ax1.legend(loc='upper left')

# P(x)
ax2 = ax1.twinx()
ax2.plot(x, Px, label='$P(x)$', color='blue')
ax2.set_ylabel('$P(x)$', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')
ax2.legend(loc='upper right')

ax1.axvline(x=x_interval_start, color='black', linestyle='--')
ax1.axvline(x=x_interval_end, color='black', linestyle='--')
ax1.text(x_interval_start + delta_x/2, 0.2, '$\delta x$', horizontalalignment='center')

plt.title('Probability Density $p(x)$ and Cumulative Distribution Function $P(x)$')
plt.show()
```
